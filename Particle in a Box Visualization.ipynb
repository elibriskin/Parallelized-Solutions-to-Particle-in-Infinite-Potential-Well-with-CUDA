```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data3.csv", header=None)
well_width = 10
fig, axs = plt.subplots(1, 2, figsize=(15, 6))
x = np.linspace(-5, 5, 100)
for index, row in enumerate(data.iterrows()):
    solution = np.array(row[1].values[0:-1])
    probability = np.abs(solution**2)
    axs[0].plot(x, solution, label=f'$\psi_{index}$')
    axs[1].plot(x, probability, label=f'$\psi_{index}$')

axs[0].legend()
axs[0].set_xlabel(f'Well Distance')
axs[0].set_ylabel('$\psi$')


axs[1].legend()
axs[1].set_xlabel(f'Well Distance')
axs[1].set_ylabel('$|\psi|^2$')
```

```python
data = pd.read_csv("data5.csv", header=None)
well_width = 10
# fig, axs = plt.subplots(1, 2, figsize=(15, 6))
x = np.linspace(-5, 5, 5000)
y = np.linspace(-5, 5, 5000)
arr_list = []
x_1, y_1 = np.meshgrid(x, y)
for index, row in enumerate(data.iterrows()):
    arr_row = np.array(row[1].values[0:-1])
    arr_list.append(arr_row)

matrix = np.vstack(arr_list)

# print(matrix)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(x_1, y_1, matrix, cmap='cool', edgecolor='blue')
ax.set_title('Surface Plot')
plt.show()
```
